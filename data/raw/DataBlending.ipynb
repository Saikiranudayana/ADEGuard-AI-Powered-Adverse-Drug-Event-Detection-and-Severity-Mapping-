```python
all_data = {}

for i in range(len(folder_name)):
    folder = folder_name[i]
    for j in range(len(files_structure[folder])):

        file_name = files_structure[folder][j]
        
        file_path = os.path.join(data_location, folder, file_name)
        
        key = f"{folder}_{file_name.replace('.csv','')}"
        print(f"Loading: {key}")
        
        all_data[key] = pd.read_csv(file_path, low_memory=False,encoding='latin1')
```

### Merging 2020 Files

```python
years = ['2020','2021','2022','2023','2024','2025']
merged_data_by_year = {}

for year in years:
    symptoms_data = None
    vaersymptoms_data = None
    vax_data = None

    for key in all_data:
        if key.startswith(year):
            if key.endswith('SYMPTOMS'):
                symptoms_data = all_data[key]
            elif key.endswith('DATA'):
                vaersymptoms_data = all_data[key]
            else:
                vax_data = all_data[key]

    if symptoms_data is not None and vaersymptoms_data is not None and vax_data is not None:
        merged_df = vaersymptoms_data.merge(symptoms_data, on='VAERS_ID').merge(vax_data, on='VAERS_ID')
        merged_data_by_year[year] = merged_df
    else:
        print(f"Missing data for {year} — cannot merge.")
```

```python
COBINED_DATA_2020 = merged_data_by_year['2020']
```

```python
import pandas as pd

data_2020 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2020.csv')
data_2021 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2021.csv')
data_2022 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2022.csv')
data_2023 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2023.csv')
data_2024 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2024.csv')
data_2025 = pd.read_csv('C:\\Users\\saiki\\OneDrive\\Desktop\\serious work\\data\\raw\\COMBINED_DATA_2025.csv')
```

```python
complete_data = [data_2020, data_2021, data_2022, data_2023, data_2024, data_2025]
years = ['2020', '2021', '2022', '2023', '2024', '2025']

for idx, data in enumerate(complete_data):
    print(f"Year: {years[idx]}")
    print(f"Shape: {data.shape}")
    print("\nNull Values:")
    
    # Create a table format for null values
    null_counts = data.isnull().sum()
    null_percent = (data.isnull().sum() / data.shape[0]) * 100
    
    # Create a DataFrame for better display
    null_table = pd.DataFrame({
        'Column Name': null_counts.index,
        'Null Count': null_counts.values,
        'Null Percentage': null_percent.values.round(2)
    })
    
    # Sort by null count in descending order for better readability
    null_table = null_table.sort_values('Null Count', ascending=False)
    
    # Display the table
    display(null_table)
    print("\n" + "="*80 + "\n")
```

```python

# First, get all unique columns across all datasets
all_columns = set()
for data in complete_data:
    all_columns.update(data.columns)
all_columns = sorted(list(all_columns))

# Create a multi-index DataFrame to hold all the information
# First level will be the column name
# Second level will be the metric (null count, null percentage, etc.)
consolidated_data = []

# Get basic info for each year
for idx, data in enumerate(complete_data):
    year = years[idx]
    
    # Get shape
    row_count = data.shape[0]
    col_count = data.shape[1]
    
    # Get null counts and percentages
    null_counts = data.isnull().sum()
    null_percent = (null_counts / row_count) * 100
    
    # Get zero counts and percentages
    zero_counts = (data == 0).sum()
    zero_percent = (zero_counts / row_count) * 100
    
    # Get unique value counts and percentages
    unique_counts = data.nunique()
    unique_percent = (unique_counts / row_count) * 100
    
    # Add to consolidated data
    year_data = {
        'Year': year,
        'Row Count': row_count,
        'Column Count': col_count,
        'Total Null Values': null_counts.sum(),
        'Total Null %': (null_counts.sum() / (row_count * col_count) * 100).round(2),
        'Total Zero Values': zero_counts.sum(),
        'Total Zero %': (zero_counts.sum() / (row_count * col_count) * 100).round(2)
    }
    consolidated_data.append(year_data)

# Create the consolidated dataframe
consolidated_df = pd.DataFrame(consolidated_data)

# Display the basic summary first
print("Summary Statistics for All Years:")
display(consolidated_df)
```
